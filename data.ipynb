```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import sys
import re

# get the input of pricing currency
coin_input = input('輸入計價幣別（輸入數字）：\n1. 全部\n2. AUD/澳幣\n3. CAD/加幣\n4. CHF/瑞士法朗\n5. CNY/人民幣\n6. EUR/歐元\n7. GBP/英鎊\n8. HKD/港幣\n9. JPY/日圓\n10. NZD/紐西蘭幣\n10. SEK/瑞典克朗\n11. SGD/新加坡幣\n11. USD/美元\n12. ZAR/南非幣\n')
if coin_input == "1":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[1]"
elif coin_input == "2":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[2]"
elif coin_input == "3":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[3]"
elif coin_input == "4":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[4]"
elif coin_input == "5":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[5]"
elif coin_input == "6":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[6]"
elif coin_input == "7":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[7]"
elif coin_input == "8":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[8]"
elif coin_input == "9":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[9]"
elif coin_input == "10":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[10]"
elif coin_input == "11":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[11]"
elif coin_input == "12":
    coin_xpath = "/html/body/div[4]/div/div/div/span/mat-option[12]"
else:
    sys.exit("輸入錯誤")

# get the input of product status
condition_input = input('輸入商品狀態（輸入數字）：\n1. 全部\n2. 未到期\n3. 已到期\n')
if condition_input == "1":
    condition_xpath = "/html/body/div[4]/div/div/div/span/mat-option[1]"
elif condition_input == "2":
    condition_xpath = "/html/body/div[4]/div/div/div/span/mat-option[2]"
elif condition_input == "3":
    condition_xpath = "/html/body/div[4]/div/div/div/span/mat-option[3]"
else:
    sys.exit("輸入錯誤")

# open the target web page
url = 'https://structurednotes.tdcc.com.tw/Snoteanc/#/BAS210'
driver = webdriver.Chrome()
driver.get(url)

# click the target pricing currency
wait = WebDriverWait(driver, 20)
coin_status = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/tdcc-snote-root/tdcc-snote-layout/div/div/tdcc-snote-bas210/tdcc-snote-frontend-bas210-shared/tdcc-snote-search-block/div/div/div/div[3]/form/div[1]/div[5]/div/tdcc-snote-select")))
coin_status.click()
wait = WebDriverWait(driver, 10)
cond = wait.until(EC.element_to_be_clickable((By.XPATH, coin_xpath)))
cond.click()

# click the target product status
wait = WebDriverWait(driver, 10)
condition = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[1]/tdcc-snote-root/tdcc-snote-layout/div/div/tdcc-snote-bas210/tdcc-snote-frontend-bas210-shared/tdcc-snote-search-block/div/div/div/div[3]/form/div[1]/div[7]/div/tdcc-snote-select/div/input")))
condition.click()
wait = WebDriverWait(driver, 10)
cond = wait.until(EC.element_to_be_clickable((By.XPATH, condition_xpath)))
cond.click()

# create a new table to store the data
table_data = []

# this loop is for find the data within January to July
for j in range(7):
    # click the calender to show the date we can choose  
    wait = WebDriverWait(driver, 10)
    calender = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/tdcc-snote-root/tdcc-snote-layout/div/div/tdcc-snote-bas210/tdcc-snote-frontend-bas210-shared/tdcc-snote-search-block/div/div/div/div[3]/form/div[1]/div[8]/div/tdcc-snote-datepicker/div/input")))
    driver.execute_script("arguments[0].scrollIntoView();", calender)
    driver.execute_script("arguments[0].click();", calender)
    # searching every month
    wait = WebDriverWait(driver, 10)
    month = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/thead/tr[1]/th[2]/select[1]")))
    month.click()
    wait = WebDriverWait(driver, 10)
    month_sel = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/thead/tr[1]/th[2]/select[1]/option[" + str(j+1) + "]")))
    month_sel.click()
    # click the first day and last day in each month  
    if j == 0:
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[1]/td[2]")))
        start.click()
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[5]/td[4]")))
        start.click()
    elif j == 1:
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[1]/td[5]")))
        start.click()
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[5]/td[5]")))
        start.click()
    elif j == 2:
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[1]/td[6]")))
        start.click()
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[6]/td[1]")))
        start.click()
    elif j == 3:
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[1]/td[2]")))
        start.click()
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[5]/td[3]")))
        start.click()
    elif j == 4:
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[1]/td[4]")))
        start.click()
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[5]/td[6]")))
        start.click()
    elif j == 5:
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[1]/td[7]")))
        start.click()
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[6]/td[1]")))
        start.click()
    else:
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[1]/td[2]")))
        start.click()
        wait = WebDriverWait(driver, 10)
        start = wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div[2]/div[1]/table/tbody/tr[5]/td[4]")))
        start.click()
    # click the comfirmation button    
    wait = WebDriverWait(driver, 10)
    confirm = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[2]/div[4]/button[2]")))
    driver.execute_script("arguments[0].scrollIntoView();", confirm)
    driver.execute_script("arguments[0].click();", confirm)
    # click the searching button
    wait = WebDriverWait(driver, 10)
    find = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/tdcc-snote-root/tdcc-snote-layout/div/div/tdcc-snote-bas210/tdcc-snote-frontend-bas210-shared/tdcc-snote-search-block/div/div/div/div[3]/form/div[2]/tdcc-snote-button/button")))
    driver.execute_script("arguments[0].scrollIntoView();", find)
    driver.execute_script("arguments[0].click();", find)
    # some months don't have data     
    try:
        element = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/tdcc-snote-root/tdcc-snote-layout/div/div/tdcc-snote-bas210/tdcc-snote-frontend-bas210-shared/div/div/div/tdcc-snote-table/div[2]/div[1]")))
        # Use JavaScript to get the text node content
        text_content = driver.execute_script("return arguments[0].childNodes[2].nodeValue;", element)
        page_number = int(re.search(r'\d+', text_content).group())
    except:
        continue
    # select data in each page  
    for i in range(page_number):
        table = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/tdcc-snote-root/tdcc-snote-layout/div/div/tdcc-snote-bas210/tdcc-snote-frontend-bas210-shared/div/div/div/tdcc-snote-table/div[1]")))
        # Extract table headers
        headers = [header.text for header in table.find_elements(By.XPATH, ".//thead//th")]
        # Extract table rows
        rows = table.find_elements(By.XPATH, ".//tbody//tr")
        # Extract data from each row
        for row in rows:
            cells = row.find_elements(By.XPATH, ".//td")
            row_data = [cell.text for cell in cells]
            table_data.append(row_data)

        if i != page_number-1:
            wait = WebDriverWait(driver, 10)
            next_page = wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/tdcc-snote-root/tdcc-snote-layout/div/div/tdcc-snote-bas210/tdcc-snote-frontend-bas210-shared/div/div/div/tdcc-snote-table/div[2]/div[2]/a[1]")))
            driver.execute_script("arguments[0].scrollIntoView();", next_page)
            driver.execute_script("arguments[0].click();", next_page)
    # search the next month     
    next_month = wait.until(EC.presence_of_element_located((By.XPATH, "//*[@id='snoteancSearch']/div/div/div[2]/tdcc-snote-button/button")))
    driver.execute_script("arguments[0].scrollIntoView();", next_month)
    driver.execute_script("arguments[0].click();", next_month)

print(table_data)
driver.quit()
```
